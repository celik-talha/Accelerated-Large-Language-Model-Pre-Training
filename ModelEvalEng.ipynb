```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch
import os

base_path = "/content/drive/MyDrive/Bitirme/Models/Gsm8k/Checkpoints"
save_base_path = "/content/drive/MyDrive/Bitirme/Models/16Bit/Checkpoints"

for step in range(6, 8):
    step_name = f"llama-1b-base-gsm8k-0.step_{step}"
    drive_model_path = os.path.join(base_path, step_name)
    save_path = os.path.join(save_base_path, step_name)

    print(f"Processing {step_name}...")

    # Model ve tokenizer'ı float16 olarak yükle
    model = AutoModelForCausalLM.from_pretrained(drive_model_path, torch_dtype=torch.float16)
    tokenizer = AutoTokenizer.from_pretrained(drive_model_path)

    # Kayıt klasörü yoksa oluştur
    os.makedirs(save_path, exist_ok=True)

    # Model ve tokenizer'ı kaydet
    model.save_pretrained(save_path, safe_serialization=True)
    tokenizer.save_pretrained(save_path)

    print(f"Saved to {save_path}")
```
