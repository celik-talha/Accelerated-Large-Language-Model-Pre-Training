```python
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM
from datasets import Dataset
import pandas as pd
from torch.utils.data import DataLoader
from tqdm import tqdm
from torch.amp import autocast

# --- Validation Dataset sadece 1 kere hazırlanır ---
val_path = "/content/drive/MyDrive/Bitirme/Datasets/validation_dataset.csv"
val_data = pd.read_csv(val_path)
val_texts = val_data["Text"].tolist()
val_dataset = Dataset.from_dict({"text": val_texts})

# --- Model yolları ---
base_path = "/content/drive/MyDrive/Bitirme/Models/Final/Cosmopedia/"
model_names = ["llama-1b-base-cosmopedia-0"] + [f"llama-1b-base-cosmopedia-slerp-{i}" for i in range(1, 10)]

# --- Her model için validation loss hesapla ---
for model_name in model_names:
    print(f"\n🔍 Evaluating: {model_name}")
    model_path = base_path + model_name

    model = AutoModelForCausalLM.from_pretrained(model_path).cuda()
    tokenizer = AutoTokenizer.from_pretrained(model_path)
    model.eval()

    def tokenize_function(examples):
      return tokenizer(
        examples["text"],
        padding="max_length",
        truncation=True,
        max_length=512,
        return_tensors="pt"
    )

    tokenized_val = val_dataset.map(tokenize_function, batched=True, remove_columns=["text"])

    val_loader = DataLoader(tokenized_val, batch_size=8, shuffle=False, collate_fn=lambda x: {
        'input_ids': torch.stack([torch.tensor(f['input_ids']) for f in x]),
        'attention_mask': torch.stack([torch.tensor(f['attention_mask']) for f in x]),
        'labels': torch.stack([torch.tensor(f['input_ids']) for f in x])
    })

    total_val_loss = 0
    total_tokens = 0

    with torch.no_grad():
        for batch in tqdm(val_loader, desc=f"Validating {model_name}"):
            batch = {k: v.cuda() for k, v in batch.items()}

            with autocast("cuda"):
                outputs = model(**batch)
                loss = outputs.loss

            token_count = batch["attention_mask"].sum().item()
            total_val_loss += loss.item() * token_count
            total_tokens += token_count

    avg_val_loss = total_val_loss / total_tokens
    print(f"✅ Validation Loss for {model_name}: {avg_val_loss:.4f}")

    # CUDA belleğini temizle
    del model
    torch.cuda.empty_cache()
```
